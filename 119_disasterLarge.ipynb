```python
import os
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments, EarlyStoppingCallback
from transformers import DataCollatorWithPadding
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import torch
from torch.utils.data import Dataset


# MPS 및 CUDA 비활성화
os.environ['PYTORCH_ENABLE_MPS_FALLBACK'] = '1'
os.environ['CUDA_VISIBLE_DEVICES'] = '-1'

# CPU로 강제 설정
device = torch.device('cpu')
torch.backends.mps.is_available = lambda: False
torch.backends.mps.is_built = lambda: False

# 데이터셋 클래스 정의
class TextDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        labels = self.labels[idx]

        encoding = self.tokenizer(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            return_token_type_ids=False,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt',
        )

        return {
            'text': text,
            'input_ids': encoding['input_ids'].flatten().to(device),  # CPU로 이동
            'attention_mask': encoding['attention_mask'].flatten().to(device),  # CPU로 이동
            'labels': torch.tensor(labels, dtype=torch.long).to(device)  # CPU로 이동
        }
# Load data
training_data_path = '/Users/ijeongbin/Desktop/dev/whisper/sst/AIHUB/data/088.위급상황_음성-음향_고도화_119_지능형_신고접수_음성_인식_데이터/3.개방데이터/1.데이터/Training/2.라벨링데이터/TL_서울.csv'
validation_data_path = '/Users/ijeongbin/Desktop/dev/whisper/sst/AIHUB/data/088.위급상황_음성-음향_고도화_119_지능형_신고접수_음성_인식_데이터/3.개방데이터/1.데이터/Training/2.라벨링데이터/TL_서울.csv'
training_df = pd.read_csv(training_data_path)
validation_df = pd.read_csv(validation_data_path)

# 데이터 준비
tokenizer = AutoTokenizer.from_pretrained("beomi/KcELECTRA-base")
label_map = {'구급': 0, '구조': 1, '기타': 2, '화재': 3}

training_df['label'] = training_df['disasterLarge'].map(label_map)
validation_df['label'] = validation_df['disasterLarge'].map(label_map)

train_texts, val_texts, train_labels, val_labels = train_test_split(
    training_df['text'].tolist(),
    training_df['label'].tolist(),
    test_size=0.1,
    random_state=42
)

train_dataset = TextDataset(train_texts, train_labels, tokenizer, max_len=128)
val_dataset = TextDataset(val_texts, val_labels, tokenizer, max_len=128)
```

```python
# 모델 정의
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-multilingual-cased", num_labels=4)
model.to(device) 

# 트레이너 설정
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,  # 에포크 수 줄임
    per_device_train_batch_size=8,  # 배치 크기 줄임
    per_device_eval_batch_size=8,
    warmup_steps=100,  # 워밍업 스텝 수 줄임
    weight_decay=0.01,  # 그대로 유지
    logging_dir='./logs',
    logging_steps=50,  # 로깅 스텝 수 늘림
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True
)

data_collator = DataCollatorWithPadding(tokenizer)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    data_collator=data_collator,
    tokenizer=tokenizer,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=3)]  # 조기 종료 설정
)

# 모델 학습
trainer.train()

# 모델 평가
predictions, labels, _ = trainer.predict(val_dataset)
preds = predictions.argmax(-1)

print(classification_report(labels, preds, target_names=label_map.keys()))
```

```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments, EarlyStoppingCallback, DataCollatorWithPadding
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import numpy as np


# 데이터 로드
training_data_path = '/Users/ijeongbin/Desktop/dev/whisper/sst/AIHUB/TL.csv'
validation_data_path = '/Users/ijeongbin/Desktop/dev/whisper/sst/AIHUB/VL.csv'
training_df = pd.read_csv(training_data_path)
validation_df = pd.read_csv(validation_data_path)

# 레이블 맵핑
label_map = {'구급': 0, '구조': 1, '기타': 2, '화재': 3}
training_df['label'] = training_df['disasterLarge'].map(label_map)
validation_df['label'] = validation_df['disasterLarge'].map(label_map)

train_sample_df = training_df.sample(frac=0.1, random_state=42)
val_sample_df = validation_df.sample(frac=0.1, random_state=42)

# 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained("beomi/KcELECTRA-base")

# 데이터셋 클래스 정의
class TextDataset(torch.utils.data.Dataset):
    def __init__(self, texts, labels, tokenizer, max_len):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        label = self.labels[idx]

        encoding = self.tokenizer(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            return_token_type_ids=False,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt',
        )

        return {
            'text': text,
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }

def preprocess_data(df, tokenizer, max_len=128):
    return TextDataset(
        texts=df['text'].tolist(),
        labels=df['label'].tolist(),
        tokenizer=tokenizer,
        max_len=max_len
    )

train_dataset = preprocess_data(training_df, tokenizer)
val_dataset = preprocess_data(validation_df, tokenizer)

# 모델 정의
model_kc = AutoModelForSequenceClassification.from_pretrained("beomi/KcELECTRA-base", num_labels=4)
model_kc.to(device)

# 트레이너 설정
training_args = TrainingArguments(
    output_dir='./results_kc',
    num_train_epochs=2,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    warmup_steps=50,
    weight_decay=0.01,
    logging_dir='./logs_kc',
    logging_steps=100,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    learning_rate=5e-5
)

data_collator = DataCollatorWithPadding(tokenizer)

trainer_kc = Trainer(
    model=model_kc,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    data_collator=data_collator,
    tokenizer=tokenizer,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=2, early_stopping_threshold=0.01)]  # 조기 종료 설정
)

# 모델 학습
trainer_kc.train()

# 모델 저장
model_save_path = './saved_model_kc'
tokenizer_save_path = './saved_tokenizer_kc'
trainer_kc.model.save_pretrained(model_save_path)
trainer_kc.tokenizer.save_pretrained(tokenizer_save_path)

# 모델 평가
predictions_kc = trainer_kc.predict(val_dataset)
preds_kc = np.argmax(predictions_kc.predictions, axis=1)
labels_kc = predictions_kc.label_ids

print(classification_report(labels_kc, preds_kc, target_names=label_map.keys()))
```
